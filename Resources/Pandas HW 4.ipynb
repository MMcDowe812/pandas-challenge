```python
import pandas as pd
from pathlib import Path
```

```python
#file to load
school_file = Path("schools_complete.csv")
student_file = Path("students_complete.csv")
```

```python
#read files
School_df=pd.read_csv(school_file)
student_df=pd.read_csv(student_file)
School_df
student_df
```

```python
# Merge files
School_data_complete = pd.merge(School_df,student_df, how = "left", on = ['school_name', 'school_name'] )
School_data_complete.head()
```

```python
#Calculate the school totals
School_count = len(School_data_complete.value_counts("school_name"))
School_count
```

```python
#Calculate the student totals
Student_total = len(School_data_complete.value_counts("Student ID"))
Student_total
```

```python
#Calculate total budget between all schools
budget_total = School_data_complete["budget"].unique().sum()
budget_total
```

```python
#Calculate average math scores
average_math = School_data_complete["math_score"].mean()
average_math
```

```python
#Calculate average reading scores
average_reading = School_data_complete["reading_score"].mean()
average_reading
```

```python
# % of students who passed math
passing_math_count = School_data_complete[(School_data_complete["math_score"]>=70)].count()["student_name"]
student_count = School_data_complete.count()["student_name"]
passing_math_pct = passing_math_count / float(student_count)*100
passing_math_pct
```

```python
#% of students who passed reading
passing_reading_count = School_data_complete[(School_data_complete["reading_score"]>=70)].count()["student_name"]
student_count = School_data_complete.count()["student_name"]
passing_reading_pct = passing_reading_count / float(student_count)*100
passing_reading_pct
```

```python
#% of students who passed both Math and reading
pass_math_reading_count = School_data_complete[(School_data_complete["math_score"]>=70) & (School_data_complete["reading_score"]>=70)].count()["student_name"]
student_count = School_data_complete.count()["student_name"]
pass_math_reading_pct = pass_math_reading_count / float(student_count)*100
pass_math_reading_pct
```

```python
# District Summary DataFrame creation
District_Summary = pd.DataFrame({"Total Schools in District": [School_count], "Total Students": [Student_total], "Total Budget": [budget_total], "Average Math Scores":[average_math],
                                "Average Reading Score": [average_reading], "% Passing Math":[passing_math_pct], "% Passing Reading":[passing_reading_pct], "% Passing Reading and Math":[pass_math_reading_pct] })
# DataFrame Formatting
District_Summary["Total Students"]=District_Summary["Total Students"].map("{:,}".format)
District_Summary["Total Budget"] = District_Summary["Total Budget"].map("${:,.2f}".format)
District_Summary["Average Math Scores"] = District_Summary["Average Math Scores"].map("{:.2f}".format)
District_Summary["Average Reading Score"] = District_Summary["Average Reading Score"].map("{:.2f}".format)
District_Summary["% Passing Math"] = District_Summary["% Passing Math"].map("{:.2f}%".format)
District_Summary["% Passing Reading"] = District_Summary["% Passing Reading"].map("{:.2f}%".format)
District_Summary["% Passing Reading and Math"] = District_Summary["% Passing Reading and Math"].map("{:.2f}%".format)
District_Summary
```

```python
# Establish the bins
spending_bins = [0,585,630,645,680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget"
school_spending_df = per_school_summary.copy()
```

```python
school_spending_df
```

```python
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"], spending_bins, labels = labels)
school_spending_df
```

```python
#  Calculate averages for the desired columns.
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Student Math Average"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Student Reading Average"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading and Math"].mean()
```

```python
# Assemble into DataFrame
spending_summary =pd.DataFrame({"Student Math Average":spending_math_scores, "Student Reading Average":spending_reading_scores.values, 
                                "% Passing Math":spending_passing_math.values, "% Passing Reading": spending_passing_reading.values, "% Overall Passing":overall_passing_spending.values})

# Display results
spending_summary
```

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

per_school_summary["Student School Size"] = pd.cut(per_school_summary["School Size"], size_bins, labels = labels)
per_school_summary
```

```python
# Calculate averages for the desired columns.
size_math_scores = per_school_summary.groupby(["Student School Size"])["Student Math Average"].mean()
size_reading_scores = per_school_summary.groupby(["Student School Size"])["Student Reading Average"].mean()
size_passing_math = per_school_summary.groupby(["Student School Size"])["% Passing Math"].mean()
size_passing_reading = per_school_summary.groupby(["Student School Size"])["% Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby(["Student School Size"])["% Passing Reading and Math"].mean()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.DataFrame({"Average Math Score":size_math_scores, "Average Reading Score":size_reading_scores.values, "% Passing Reading": size_passing_math.values,
                            "% Passing Reading": size_passing_reading.values, "% Passing Reading and Math": size_overall_passing.values})

# Display results
size_summary
```

```python
# Group the per_school_summary DataFrame by "School Type" and average the results.
average_math_score_by_type = per_school_summary.groupby(["School Types"])["Student Math Average"].mean()
average_reading_score_by_type = per_school_summary.groupby(["School Types"])["Student Reading Average"].mean()
average_percent_passing_math_by_type = per_school_summary.groupby(["School Types"])["% Passing Math"].mean()
average_percent_passing_reading_by_type = per_school_summary.groupby(["School Types"])["% Passing Reading"].mean()
average_percent_overall_passing_by_type = per_school_summary.groupby(["School Types"])["% Passing Reading and Math"].mean()
```

```python
type_summary = pd.DataFrame({"Average Math Score":average_math_score_by_type, "Average Reading Score":average_reading_score_by_type.values, "% Passing Reading": average_percent_passing_math_by_type.values,
                            "% Passing Reading": average_percent_passing_reading_by_type.values, "% Passing Reading and Math": average_percent_overall_passing_by_type.values})
type_summary
```
